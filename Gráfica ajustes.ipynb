```python
import numpy as np
%pylab inline
```

```python
x=np.array([(-2,0),(-1,-1),(0,1),(1,0),(2,-2),(4,3)])
scatter(x[:,0],x[:,1])
n=linspace(-3,5,1000)
lin=(5*n/14)-1/14
cuad=(((3*n**2)/14)-(n/14)-5/7)
plot(n,lin,color="k")
plot (n,cuad, color="g")
axis("equal")
grid(True)
ylabel("y")
xlabel("x")
title("Punto a")
legend(("Ajuste lineal","Ajuste cuadrático","Puntos"))
axvline(0,0,1,linestyle='--',color='k')
axhline(0,0,1,linestyle='--',color='k')
```

```python
y=np.array([(0,1),(-1,6),(1,0),(2,3)])
scatter(y[:,0],y[:,1])
k=linspace(-1.1,2.5,1000)
lin1=k+3
cuad1=((2*k**2)-(3*k)+1)
plot(k,lin1,color="k")
plot (k,cuad1, color="g")
axis("equal")
grid(True)
ylabel("y")
xlabel("x")
title("Punto b")
legend(("Ajuste lineal","Ajuste cuadrático","Puntos"))
axvline(0,0,1,linestyle='--',color='k')
axhline(0,0,1,linestyle='--',color='k')
```
